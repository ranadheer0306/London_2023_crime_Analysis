```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crimes = pd.read_csv("/content/drive/MyDrive/London Crimes Project/London_2023_Crimes.csv")
```

```python
crimes.head()
```

```python
crimes.tail()
```

```python
crimes['Reported by'].count()
# There are total 9298 street crimes reported in london from Janunary 2023 to December 2023
```

```python
crimes = crimes.drop('Context', axis = 1)
```

```python
sns.barplot(y=crimes['Crime type'].value_counts().index,
            x=crimes['Crime type'].value_counts(),
            order=crimes['Crime type'].value_counts().index)
plt.ylabel('Crime Type')
plt.title('Number of crimes happened in London Streets')
plt.xlabel('Count of Crimes')
plt.show()
# The below plot shows the count of different crimes reported
# we can understand the more than 60% of the crimes are related to theft, voilence and sexual offences
```

```python
crimes['Last outcome category'].value_counts()
```

```python
sns.barplot(y=crimes['Last outcome category'].value_counts().index,
            x=crimes['Last outcome category'].value_counts(),
            order=crimes['Last outcome category'].value_counts().index)
plt.title('Result of Crime')
plt.xlabel('Count')
plt.ylabel('Last outcome category')
plt.show()

# Outcome of the most cases in 2023
```

```python
crimes[crimes['Crime type'] == "Other theft"]['Last outcome category'].value_counts()
#Around 90% of the theft cases in london that happened in 2023 resulted as No suspect found
```

```python
sns.barplot(y=crimes[crimes['Crime type'] == "Other theft"]['Last outcome category'].value_counts().index,
            x=crimes[crimes['Crime type'] == "Other theft"]['Last outcome category'].value_counts(),
            order=crimes[crimes['Crime type'] == "Other theft"]['Last outcome category'].value_counts().index)
plt.title('Result of Theft Crime')
plt.xlabel('Count')
plt.ylabel('Last outcome category')
plt.show()
```

```python
#comparision of crimes by month
sns.barplot(crimes['Month'].value_counts())
plt.xticks(rotation = 90)
plt.ylabel("Total street crimes")
plt.xlabel("Month")
plt.title("Number of crimes in each Month in 2023")
plt.show()
```
